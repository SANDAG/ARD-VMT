```python
emfac = pl.concat([emfac2017, emfac2021, emfac2025]).with_columns(
    pl.col("Calendar Year").cast(pl.String).str.to_date(format="%Y"),
)
emfac.write_parquet("../../data/clean/emfac/emfac.parquet")
```

```python
(
    emfac.group_by(["EMFAC Model", "Calendar Year"])
    .sum()
    .select(
        "EMFAC Model",
        "Calendar Year",
        "Total VMT",
    )
    .plot.line(
        x="Calendar Year",
        y="Total VMT",
        color="EMFAC Model",
    )
)
```
